```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1=pd.read_csv('/content/RS_Session_260_AU_2349_A_to_B.csv')
```

```python
df1.head()
```

```python

df1.info()
```

```python
df1.shape
```

```python
df1["Percent_Electric"] = (df1["Electric"]/df1["Total"])*100
```

```python

df1["Non-Electric to Electric Ratio"] = df1["Non-electric"]/df1["Electric"]
```

```python
df1.head(2)
```

```python
# Set a clean background style
sns.set_style("whitegrid")

# Create a large figure with a visually appealing color scheme
plt.figure(figsize=(45, 20))
sns.barplot(x=df1["State/UT"], y=df1["Percent_Electric"], palette="Set3")

# Customize labels and title (optional)
plt.xlabel("State/UT")  # Adjust label if needed (e.g., "State/UT")
plt.ylabel("Percent Electric")
plt.title("Percent Electric by State (Large Format)")

# Rotate x-axis labels for readability if needed
plt.xticks(rotation=45, ha="right")

# Tighten layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(45, 20))
sns.barplot(
    x=df1["State/UT"],
    y=df1["Non-Electric to Electric Ratio"],
    palette="Set3"  # Use a visually appealing color palette from Seaborn
)

# Customize labels and title (optional)
plt.xlabel("State/UT")
plt.ylabel("Non-Electric to Electric Ratio")
plt.title("Non-Electric to Electric Ratio by State/UT (Large Format)")

# Display the plot
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust spacing to prevent overlapping elements
plt.show()
```

```python
df=pd.read_csv('/content/car_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,8))

fig.suptitle("Relationship of car prices and fuel types and majority people purchase capacity")

axes[0].set_title("Car Prices with respect to Fuel")
ax1 = sns.barplot(x = list(df['Fuel_Type']),y = list(df['Present_Price']),data=df,ax=axes[0])
ax1.set(xlabel = "Fuel Types",ylabel = "Prices in Lakhs")

dict_df = dict(df['Fuel_Type'].value_counts())
plt.subplot(122)
plt.pie(list(dict_df.values()),labels=list(dict_df.keys()),autopct='%.0f%%')
plt.title("Share of the cars with particular Fuel Type")
plt.xlabel("Percentage share of cars with particular fuel type")

plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,8))

fig.suptitle("Relationship of car prices and tranasmission types and majority people purchase capacity")

axes[0].set_title("Car Prices with respect to transmission")
ax = sns.barplot(x = list(df['Transmission']),y = list(df['Present_Price']),data=df,ax=axes[0])
ax.set(xlabel = "Transmission Types",ylabel = "Prices in Lakhs")


dict_df = dict(df['Transmission'].value_counts())
plt.subplot(122)
plt.pie(list(dict_df.values()),labels=list(dict_df.keys()),autopct='%.0f%%')
plt.title("Share of the cars with particular transmission Type")
plt.xlabel("Percentage share of cars with particular transmission type")

plt.show()
```
