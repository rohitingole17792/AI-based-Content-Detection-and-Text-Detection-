#Data Analysis & Visualization

```python
canceled_per =df['is_canceled'].value_counts(normalize=True) #value_counts is a funcation ,[''] how many times present in coloumn # (normalize=True) persentage Return
canceled_per
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancel ','cancel'],df['is_canceled'].value_counts(),  edgecolor= 'k',width=0.7)
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel', hue='is_canceled' , data=df , palette ='Blues')
# legend_lables,_ = ax1. get_legend_handles_lables()
# ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation of Different hotel',size=20)
plt.xlabel('Hotel')
plt.ylabel('No of Reservation')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel['is_canceled'].value_counts(normalize=True).plot.bar()
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel['is_canceled'].value_counts(normalize=True).plot.bar()
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Avarage Daily Rate In City and Resort Hotel' , fontsize=35)
plt.plot(resort_hotel.index,resort_hotel['adr'] , label= 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'] , label= 'City Hotel')
# plt.leagend(fontsize= 20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
ax1 = sns.countplot(x = 'month', hue='is_canceled' , data=df ,palette='bright')
plt.title('Reservation status month',size=30)
plt.xlabel('month')
plt.ylabel('no of reservation')
plt.legend(['not_cancelled','cancelled'])
plt.show()
```

```python
import seaborn as sns
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (17,8))
sns.countplot(x = 'month',hue = 'is_canceled', data = df, palette = "BuPu")
plt.title("Reservation Status per month", fontsize = 20)
plt.legend(["not cancelled", "cancelled"])
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', size=20)

sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
cancelled_reservations = df[df['is_canceled'] == 1]
not_cancelled_reservations = df[df['is_canceled'] == 0]
```

```python
top_10_countries = (cancelled_reservations['country'].value_counts()[:10])
plt.figure(figsize = (20,8))
plt.title("Top 10 country hotels with cancelled reservations")
plt.pie(top_10_countries,labels = top_10_countries.index, autopct = '%.2f', colors=sns.color_palette('BuPu'),
)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
market_segment = cancelled_reservations['market_segment'].value_counts(normalize = True)
market_segment.sort_values()
```

```python
plt.figure(figsize = (20,8))
plt.title("Market Segment of the hotel", fontsize = 20)
explode = (0,0,0,0.2,0.3,0.4,0.9)
plt.pie(market_segment, labels = market_segment.index, autopct = '%.2f', colors=sns.color_palette('BuPu'), explode = explode)
plt.show()
```
