```python
import pandas as pd
import numpy as np
import seaborn as sas
```

```python
df=pd.read_csv('/content/iris.csv')
```

#Binary Classification

```python
df['species']!='setosa'
```

```python
df=df[df['species']!='setosa']
```

```python
df.head()
```

```python
df['species'].map({'versicolor':0,'virginica':1})
```

```python
df['species']=df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.tail()
```

```python
X = df.iloc[:, :-1]  # Selecting all rows and all columns except the last one(Independent)
y = df.iloc[:,-1]    #selects the last column of the DataFrame(Dependent)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from inspect import Parameter
from sklearn.model_selection import GridSearchCV
Parameter= {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,8,9,10,15,18,20,30,45,48,50],'max_iter':[100,200,300,400,500,600]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=Parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
#Predict

y_pred=classifier_regressor.predict(X_test)
```

```python
#accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
